## Two-story shear frame with Bouc-Wen hysteretic links
#### Example main file to evaluate the response of the frame under a ground motion excitation signal

Please cite as:
K. Vlachas, K. Tatsis, K. Agathos, A. Brink, and E. Chatzi,
A local basis approximation approach for nonlinearparametric model order reduction,
Journal of Sound and Vibration, vol. 502, p. 116055, 2021.


```python
#Import necessary packages
import numpy as np
from matplotlib import pyplot as plt
import numpy.linalg as ln
from scipy.sparse import coo_matrix
from scipy.sparse.linalg import eigs
import scipy.io
import random
```

#### Input File

Currently only an example configuration is provided, corresponding to the *InputFileLinksAll* input file of the MATLAB implementation. This assumes that hysteretic links are located in all horizontal beams, on the top of the basement columns and on the top and bottom of the first-floor columns of the two-story frame. The variables defined are explained in detail in the corresponding *ReadMe* file.

#### Material Definition 

Instansiate the material class. The class object assumes as an input the material properties, namely the Young modulus of elasticity ($ Eyoung $), the poisson ration ($ nee $) and the density ($ rho $). It also defines the cross section properties for the beam elements. Currently only an example cross section is provided for reference.


```python
class Material:
    def __init__(self, Eyoung=210e9, nee=0.30, rho=8000):
        self.mat_properties = np.array([[Eyoung, nee, rho]])

        # Steel HEA 200 cross section
        A, I1, I2, I3 = 5383e-06, 204.3e-09, 36.92e-06, 13.36e-06

        self.cross_sections = np.array([[A, I1, I2, I3]])
```

#### Bouc-Wen Model

Instansiate the class for the Bouc-Wen model of the hysteretic links. The class object assumes as input the parameters of the Bouc-Wen model governing the behavior of the hysteretic links. The notation follows the provided * ReadMe * file and the description. The corresponding notation is $ bw_a=a$, $ Alpha=A $, $ N=w $, $ Beta=\beta$, $ Gamma=\gamma $,$ deltav=dv $, $ deltan=dnu $, $bw_k$ = $ k $ 

```python
class BoucWenModel:
    def __init__(
        self,
        Beta,
        Gamma,
        bwa=0.1,
        AmpBW=1.0,
        Alpha=1.0,
        N=1,
        deltav=0,
        deltan=0,
        bw_k=1e08,
    ):
        self.bwa, self.bw_k = bwa, bw_k
        self.Beta, self.Gamma, self.deltav, self.deltan = Beta, Gamma, deltav, deltan
        self.AmpBW, self.Alpha, self.N = AmpBW, Alpha, N
        self.Properties = [bwa, bw_k, Alpha, Beta, Gamma, N, AmpBW, deltav, deltan]

    def evaluateBoucWen(self, U, DofsLoop, HistUm, HistR, HistZeta, HistE):

        Um = U[DofsLoop[:, 0].astype(int)] - U[DofsLoop[:, 1].astype(int)]
        DU = Um - HistUm

        # Strength deterioration
        Energy = HistE
        ve = 1 + self.deltav * Energy
        # Stiffness degradation
        ne = 1 + self.deltan * Energy

        zetaprior = HistZeta
        kbw = (
            self.Alpha
            - ve
            * (self.Beta * np.sign(zetaprior) * np.sign(DU) + self.Gamma)
            * np.absolute(zetaprior) ** (self.N)
        ) / ne

        # Force terms
        LinearTerm = self.bwa * self.bw_k * DU
        dzeta = kbw * DU
        HystereticTerm = (1 - self.bwa) * self.bw_k * dzeta

        # Restoring force
        r = HistR + self.AmpBW * HystereticTerm + LinearTerm
        k = self.bwa * self.bw_k + (1 - self.bwa) * self.bw_k * kbw

        # Update History
        HistR = r
        HistUM = HistUm + DU
        HistZeta = zetaprior + dzeta
        HistE = HistE + dzeta * DU

        d = dict()
        d["HistR"] = HistR
        d["HistUm"] = HistUM
        d["HistZeta"] = HistZeta
        d["HistE"] = HistE
        d["Rs"] = r
        d["Ks"] = k
        return d
```

#### Frame Assembly Definition 

Instansiate the assembly class. The class contains all functions necessary to assemble the frame and the corresponding system matrices, including mass, stifness and external excitation. Inside this class damping is also defined. 


```python
class Assembly:
    def __init__(self, InputFile, Material, SimulationParameters, BoucWenModel):
        self.InputFile = InputFile
        self.Material = Material
        self.BoucWenModel = BoucWenModel
        self.SimulationParameters = SimulationParameters

        DofsLoop = Assembly.GetDofsLoop(
            InputFile.nl_link_elements, InputFile.nl_link_flags, BoucWenModel.Properties
        )
        self.DofsLoop = DofsLoop
        
        # Bouc-Wen history array initialization
        d = dict()
        d["HistR"] = np.zeros(shape=(DofsLoop.shape[0]))
        d["HistUm"] = np.zeros(shape=(DofsLoop.shape[0]))
        d["HistZeta"] = np.zeros(shape=(DofsLoop.shape[0]))
        d["HistE"] = np.zeros(shape=(DofsLoop.shape[0]))

        # Initial conditions
        ndofs = 6 * InputFile.Nodes.shape[0]
        nt = np.shape(SimulationParameters.SynthesizedAccelerogram)[0]
        DisplacementsU = np.zeros(shape=(ndofs, nt))
        VelocitiesV = np.zeros(shape=(ndofs, nt))
        AccelerationsA = np.zeros(shape=(ndofs, nt))
        d["HystereticCurvesR"] = np.zeros(shape=(DofsLoop.shape[0], nt))
        d["HystereticCurvesU"] = np.zeros(shape=(DofsLoop.shape[0], nt))
        self.HistDict = d
        
        if SimulationParameters.ICs is not None:
            VelocitiesV[:, 0] = SimulationParameters.ICs["Velocities"][:, 0]
            DisplacementsU[:, 0] = SimulationParameters.ICs["Disps"][:, 0]

        self.Displacements = DisplacementsU
        self.Velocities = VelocitiesV
        self.Accelerations = AccelerationsA
        

        # Assemble system matrices (stiffness,mass, internal forces)
        globalMatrices = Assembly.assemble(self)
        ndofs = globalMatrices["Kstiff"].shape[0]
        FinalMatrices = self.applyBCs(
            InputFile,
            globalMatrices["Kstiff"],
            globalMatrices["Mmass"],
            globalMatrices["fint"],
        )

        # Damping treatment
        if SimulationParameters.ICs["zeta"] is not None:
            Kstiff = globalMatrices["Kstiff"]
            Mmass = globalMatrices["Mmass"]
            FinalMs = self.applyBCsdelete(InputFile, Kstiff, Mmass)

            test = np.diag(Mmass)
            test = test[test > 0]
            m = np.min(test) / 1000
            for i in range(np.shape(Mmass)[0]):
                if Mmass[i, i] == 0:
                    Mmass[i, i] = m

            DampingParams = self.RayleighDamping(
                FinalMs["Stiff"],
                FinalMs["Mass"],
                SimulationParameters.ICs["zeta"],
                SimulationParameters.ICs["OmegaIndexes"],
            )
            alpha, beta = DampingParams["alpha"], DampingParams["beta"]
        else:
            alpha, beta = 0.05, 0

        Crayleigh = alpha * FinalMatrices["Mmass"] + beta * FinalMatrices["Kstiff"]

        FinalMatrices["Crayleigh"] = Crayleigh
        self.FinalMatrices = FinalMatrices


        # Assembly of excitation matrix 
        zs = Assembly.AssembleSandZ(SimulationParameters, InputFile, Mmass)
        zs[0] = np.reshape(zs[0],(len(zs[0]),1))

        Fmatrix = np.zeros(shape=(ndofs, nt))
        Fmatrix[:, 0 : np.shape(zs[0])[0]] = zs[1].dot(zs[0].T)

        self.Rmatrix = Fmatrix
        self.pos = np.argmax(zs[0])
        self.fnorm = np.linalg.norm(Fmatrix[:, self.pos])
        
    
    #Function that assembles system matrices
    def assemble(self, DispsU=None, HistDict=None):
        Mesh, Material, BoucWenModel = self.InputFile, self.Material, self.BoucWenModel
        ndof = 6 * Mesh.Nodes.shape[0]
        
        #Initialization
        Kstiff, Mmass = np.zeros(shape=(ndof, ndof)), np.zeros(shape=(ndof, ndof))
        fint = np.zeros(shape=(ndof, 1))

        condition = False
        if DispsU is None:
            condition = True
            DispsU = np.zeros(shape=(ndof))
            Mesh.Kes, Mesh.Mes, Mesh.ElDofs = [], [], []

        beam_mat_props = Material.mat_properties
        beam_cross = Material.cross_sections
        
        ## Loop over beam elements
        for e in range(Mesh.BeamElements.shape[0]):
            el = Mesh.BeamElements[e]
            #Evaluate element nodes and coordinates
            beam_nodes = el[1:] 
            beam_nodes = beam_nodes.astype(int)
            beam_Coords = Mesh.Nodes[beam_nodes - 1, :]
            
            #Assemble local element matrices
            if condition:
                beam_dofs = Assembly.elementglobaldofs(beam_nodes)
                beam_dofs = beam_dofs.astype(int)
                returnfs = Assembly.assembleBeamElement(
                    beam_mat_props, beam_cross, beam_Coords
                )
                Ke, Me = returnfs[0:2]
                Mesh.Kes.append(Ke)
                Mesh.Mes.append(Me)
                Mesh.ElDofs.append(beam_dofs)
            else:
                beam_dofs = Mesh.ElDofs[e]
                Ke, Me = Mesh.Kes[e], Mesh.Mes[e]
                
            #Assemble local element matrices to global system matrices
            
            Ue = DispsU[beam_dofs]

            Kstiff[np.ix_(beam_dofs, beam_dofs)] = (
                Kstiff[np.ix_(beam_dofs, beam_dofs)] + Ke
            )
            Mmass[np.ix_(beam_dofs, beam_dofs)] = (
                Mmass[np.ix_(beam_dofs, beam_dofs)] + Me
            )

            fint[beam_dofs, 0] = fint[beam_dofs, 0] + Ke.dot(Ue)

        ## Assembly of nonlinear links
        
        DofsLoop = self.DofsLoop
        if HistDict is None:
            HistDict = self.HistDict
            
        #Evaluate bouc-wen links response
        returndict = BoucWenModel.evaluateBoucWen(
            DispsU,
            DofsLoop,
            HistDict["HistUm"],
            HistDict["HistR"],
            HistDict["HistZeta"],
            HistDict["HistE"],
        )

        Rs, Ks = returndict["Rs"], returndict["Ks"]
        
        #Assemble response in global matrices
        rows = np.concatenate((DofsLoop[:, 0], DofsLoop[:, 1]), axis=None)
        cols = np.zeros(shape=(2 * DofsLoop[:, 0].shape[0]))
        values = np.concatenate((Rs, -Rs), axis=None)
        Fnl = coo_matrix((values, (rows, cols)), shape=(ndof, 1))

        fint = fint + Fnl.toarray()

        rows = np.concatenate(
            (DofsLoop[:, 0], DofsLoop[:, 0], DofsLoop[:, 1], DofsLoop[:, 1]), axis=None
        )
        cols = np.concatenate(
            (DofsLoop[:, 0], DofsLoop[:, 1], DofsLoop[:, 0], DofsLoop[:, 1]), axis=None
        )
        values = np.concatenate((Ks, -Ks, -Ks, Ks), axis=None)
        Kbw = coo_matrix((values, (rows, cols)), shape=(ndof, ndof))

        Kstiff = Kstiff + Kbw.toarray()

        globalreturns = dict()
        globalreturns["Kstiff"] = Kstiff
        globalreturns["Mmass"] = Mmass
        globalreturns["fint"] = fint
        globalreturns["Hist"] = returndict

        return globalreturns
    
    #Function to apply boundary conditions
    def applyBCs(self, InputFile, Kstiff, Mmass, fint):
        ndofs = 6 * InputFile.Nodes.shape[0]

        nodal_disps = InputFile.NodalDisplacements
        bc_dofs = []
        bc_dofs_nz = []
        bc_values = []

        for i in range(nodal_disps.shape[0]):
            node = nodal_disps[i, 0]
            dofs = np.arange(6 * (node - 1), 6 * node)
            dofs_act = dofs[np.where(nodal_disps[i, 1:7] != 0)[0]]
            bc_dofs.extend(dofs_act)
            nz = np.where(nodal_disps[i, 7:] != 0)[0]
            if nz.size > 0:
                bc_dofs_nz.extend(dofs[nz])
                bc_values.extend(nodal_disps[i, nz + 7])

        bc_dofs_nzA = np.array(bc_dofs_nz)
        bc_valuesA = np.array(bc_values)

        if bc_dofs_nzA.size > 0:
            fbc = Kstiff[:, bc_dofs_nzA].dot(bc_valuesA)
            # f = f - fbc

        bcdofs = np.array(bc_dofs, dtype=int)

        Kstiff[bcdofs, :] = 0
        Kstiff[:, bcdofs] = 0
        Mmass[bcdofs, :] = 0
        Mmass[:, bcdofs] = 0
        # f[bcdofs, 0] = 0
        fint[bcdofs, 0] = 0

        if bc_dofs_nzA.size > 0:
            # f[bc_dofs_nzA] = bc_valuesA
            fint[bc_dofs_nzA] = 0

        # diag_el = (bcdofs-1)*(ndofs+1)+1
        Kstiff[bcdofs, bcdofs] = 1
        Mmass[bcdofs, bcdofs] = 1
        freedofs = np.arange(ndofs)
        freedofs = np.delete(freedofs, bcdofs)

        finalms = {}
        finalms["Kstiff"] = Kstiff
        finalms["Mmass"] = Mmass
        finalms["fint"] = fint
        finalms["freedofs"] = freedofs

        return finalms

    def applyBCsdelete(self, InputFile, Kstiff, Mmass):
        # Mesh=self.Mesh
        # ndofs = 6*Mesh.Nodes.shape[0]

        nodal_disps = InputFile.NodalDisplacements
        bc_dofs = []
        bc_dofs_nz = []
        bc_values = []

        for i in range(nodal_disps.shape[0]):
            node = nodal_disps[i, 0]
            dofs = np.arange(6 * (node - 1), 6 * node)
            dofs_act = dofs[np.where(nodal_disps[i, 1:7] != 0)[0]]
            bc_dofs.extend(dofs_act)
            nz = np.where(nodal_disps[i, 7:] != 0)[0]
            if nz.size > 0:
                bc_dofs_nz.extend(dofs[nz])
                bc_values.extend(nodal_disps[i, nz + 7])

        bc_dofs_nzA = np.array(bc_dofs_nz)
        bc_valuesA = np.array(bc_values)

        # if bc_dofs_nzA.size > 0:
        #    fbc = Kstiff[:, bc_dofs_nzA].dot(bc_valuesA)
        #    f = f - fbc
        #    f[bc_dofs_nzA] = bc_valuesA
        #    fint[bc_dofs_nzA] = 0
        #    #nodal_disps[bc_dofs_nzA]=0

        bcdofs = np.array(bc_dofs, dtype=int)

        Kstifft = np.delete(Kstiff, bcdofs, 0)
        Kstiffd = np.delete(Kstifft, bcdofs, 1)
        Mmasst = np.delete(Mmass, bcdofs, 0)
        Mmassd = np.delete(Mmasst, bcdofs, 1)
        # fd = np.delete(f, bcdofs,0)
        # fintd = np.delete(fint, bcdofs,0)

        returns = {}
        returns["Stiff"] = Kstiffd
        returns["Mass"] = Mmassd

        return returns
    
    #Function to evaluate Rayleigh damping coefficients
    @staticmethod
    def RayleighDamping(Kstiff, Mmass, zetas=[0.01, 0.01], OmegasS=[1, 2]):
        eigvals, eigvecs = eigs(Kstiff, k=10, M=Mmass, sigma=0)
        Omegas = np.sqrt(eigvals.real)
        Omega2 = Omegas[OmegasS[1]]
        Omega1 = Omegas[OmegasS[0]]

        if len(zetas) == 1:
            alpha, beta = 2 * Omega1 * zetas[0], 0
        elif len(zetas) == 2:
            beta = (2 * Omega2 * zetas[1] - 2 * Omega1 * zetas[0]) / (
                Omega2 ** 2 - Omega1 ** 2
            )
            alpha = 2 * zetas[0] * Omega1 - beta * Omega1 ** 2

        dampreturns = dict()
        dampreturns["alpha"] = alpha
        dampreturns["beta"] = beta

        return dampreturns
    
    ## Functions necessary for the assembly of the local system matrices
    
    @staticmethod
    def assembleBeamElement(beam_mat_props, beam_cross, beam_Coords):
        Eyoung, nee, rho = beam_mat_props[0]
        L = ln.norm(beam_Coords[0, :] - beam_Coords[1, :])
        MatrixT = Assembly.beam_transform_matrix(beam_Coords[0, :], beam_Coords[1, :])

        A, I1, I2, I3 = beam_cross[0]
        f1, f2 = np.zeros(shape=(3)), np.zeros(shape=(3))
        elem_matrices = Assembly.beam_mass_stiffness(
            L, A, I1, I2, I3, Eyoung, nee, rho, f1, f2
        )
        M, K, f = elem_matrices[0], elem_matrices[1], elem_matrices[2]
        Ke = (MatrixT.T.dot(K)).dot(MatrixT)
        Me = (MatrixT.T.dot(M)).dot(MatrixT)
        fe = MatrixT.T.dot(f)

        return [Ke, Me, fe]

    @staticmethod
    def elementglobaldofs(nodes):
        dofs = np.zeros(shape=(len(nodes) * 6))
        dofs[0 : len(nodes) * 6 : 6] = (nodes - 1) * 6
        dofs[1 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 1
        dofs[2 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 2
        dofs[3 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 3
        dofs[4 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 4
        dofs[5 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 5
        return dofs

    @staticmethod
    def beam_transform_matrix(node1, node2):
        ex = np.array([1, 0, 0])
        ey = np.array([0, 1, 0])
        ez = np.array([0, 0, 1])

        epsilons = Assembly.beam_local_vectors(node1, node2)
        e1, e2, e3 = epsilons[0], epsilons[1], epsilons[2]

        Ti = np.array(
            [
                [np.dot(e1, ex), np.dot(e1, ey), np.dot(e1, ez)],
                [np.dot(e2, ex), np.dot(e2, ey), np.dot(e2, ez)],
                [np.dot(e3, ex), np.dot(e3, ey), np.dot(e3, ez)],
            ]
        )

        MatrixT = np.zeros(shape=(12, 12))
        MatrixT[0:3, 0:3] = Ti
        MatrixT[3:6, 3:6] = Ti
        MatrixT[6:9, 6:9] = Ti
        MatrixT[9:12, 9:12] = Ti

        # MatrixT = np.array([
        #    [Ti, np.zeros(shape=(3,3)), np.zeros(shape=(3,3)), np.zeros(shape=(3,3))],
        #    [np.zeros(shape=(3,3)), Ti, np.zeros(shape=(3,3)), np.zeros(shape=(3,3))],
        #    [np.zeros(shape=(3,3)), np.zeros(shape=(3,3)), Ti, np.zeros(shape=(3,3))],
        #    [np.zeros(shape=(3,3)), np.zeros(shape=(3,3)), Ti]
        # ])

        return MatrixT

    @staticmethod
    def AssembleSandZ(SimulationParameters, InputFile, Mass):
        Zacceler = SimulationParameters.SynthesizedAccelerogram
        ndofs = 6 * InputFile.Nodes.shape[0]
        originalnodes = InputFile.Nodes.shape[0] - InputFile.nl_link_elements.shape[0]
        angle = SimulationParameters.angle
        loaddofsx = (np.arange(originalnodes)) * 6
        loaddofsy = (np.arange(originalnodes)) * 6 + 1
        # loaddofsz = (np.arange(originalnodes)) * 6 + 2

        alldofsx = (np.arange(ndofs))[0::6]
        alldofsy = (np.arange(ndofs))[1::6]
        alldofsz = (np.arange(ndofs))[2::6]

        if type(angle) == float:
            Loaddofs = np.sort(np.hstack((loaddofsx, loaddofsy)))
            Alldofs = np.sort(np.hstack((alldofsx, alldofsy, alldofsz)))
            Mlumped = Assembly.GetLumpedMass(Mass, Alldofs)
            Mlumped = np.diag(Mlumped)
            S = np.zeros(shape=(ndofs, 1))
            S[Loaddofs, 0] = Mlumped[Loaddofs]
            indexes = Loaddofs[::2]
            S[indexes, 0] = S[indexes, 0] * np.cos(angle)
            indexes = Loaddofs[1::2]
            S[indexes, 0] = S[indexes, 0] * np.sin(angle)
        elif len(angle) == 3:
            S = 0  # Needs update

        return [Zacceler, S]

    @staticmethod
    def GetLumpedMass(Mass, translational_dofs):
        diagonal_coeff = np.diag(Mass)
        total_mass = np.sum(Mass)
        coeff_s = np.sum(diagonal_coeff[translational_dofs])
        lumped_diag = diagonal_coeff * total_mass / coeff_s
        Mlumped = np.diag(lumped_diag)
        return Mlumped

    @staticmethod
    def beam_local_vectors(node1, node2):
        ex = np.array([1, 0, 0])
        # ey=np.array([0,1,0])
        ez = np.array([0, 0, 1])

        e1 = (node2 - node1) / ln.norm(node2 - node1)
        is_all_zero = np.all((np.cross(e1, ez) == 0))
        if is_all_zero:
            e2 = np.cross(e1, ex) / ln.norm(np.cross(e1, ex))
        else:
            e2 = np.cross(e1, -ez) / ln.norm(np.cross(e1, -ez))

        e3 = np.cross(e1, e2)
        return [e1, e2, e3]

    @staticmethod
    def beam_mass_stiffness(L, A, I1, I2, I3, E, nee, rho, f1, f2):
        G = E / (2 * (1 + nee))

        Ka = np.array(
            [
                [E * A / L, 0, 0, 0, 0, 0],
                [0, 12 * E * I3 / (L ** 3), 0, 0, 0, 6 * E * I3 / (L ** 2)],
                [0, 0, 12 * E * I2 / (L ** 3), 0, -6 * E * I2 / (L ** 2), 0],
                [0, 0, 0, G * I1 / L, 0, 0],
                [0, 0, -6 * E * I2 / (L ** 2), 0, 4 * E * I2 / L, 0],
                [0, 6 * E * I3 / (L ** 2), 0, 0, 0, 4 * E * I3 / L],
            ]
        )

        Kb = np.array(
            [
                [-E * A / L, 0, 0, 0, 0, 0],
                [0, -12 * E * I3 / (L ** 3), 0, 0, 0, 6 * E * I3 / (L ** 2)],
                [0, 0, -12 * E * I2 / (L ** 3), 0, -6 * E * I2 / (L ** 2), 0],
                [0, 0, 0, -G * I1 / L, 0, 0],
                [0, 0, 6 * E * I2 / (L ** 2), 0, 2 * E * I2 / L, 0],
                [0, -6 * E * I3 / (L ** 2), 0, 0, 0, 2 * E * I3 / L],
            ]
        )

        Kc = np.array(
            [
                [E * A / L, 0, 0, 0, 0, 0],
                [0, 12 * E * I3 / (L ** 3), 0, 0, 0, -6 * E * I3 / (L ** 2)],
                [0, 0, 12 * E * I2 / (L ** 3), 0, 6 * E * I2 / (L ** 2), 0],
                [0, 0, 0, G * I1 / L, 0, 0],
                [0, 0, 6 * E * I2 / (L ** 2), 0, 4 * E * I2 / L, 0],
                [0, -6 * E * I3 / (L ** 2), 0, 0, 0, 4 * E * I3 / L],
            ]
        )

        K1 = np.zeros(shape=(12, 12))
        K1[0:6, 0:6] = Ka
        K1[0:6, 6:] = Kb
        K1[6:, 0:6] = Kb.T
        K1[6:, 6:] = Kc

        Ma = np.array(
            [
                [1 / 3, 0, 0, 0, 0, 0],
                [0, 13 / 35, 0, 0, 0, 11 * L / 210],
                [0, 0, 13 / 35, 0, -11 * L / 210, 0],
                [0, 0, 0, I1 / (3 * A), 0, 0],
                [0, 0, -11 * L / 210, 0, L ** 2 / 105, 0],
                [0, 11 * L / 210, 0, 0, 0, L ** 2 / 105],
            ]
        )

        Ma = Ma * rho * A * L

        Mb = np.array(
            [
                [1 / 6, 0, 0, 0, 0, 0],
                [0, 9 / 70, 0, 0, 0, -13 * L / 420],
                [0, 0, 9 / 70, 0, 13 * L / 420, 0],
                [0, 0, 0, -I1 / (6 * A), 0, 0],
                [0, 0, -13 * L / 420, 0, -(L ** 2) / 140, 0],
                [0, 13 * L / 420, 0, 0, 0, -(L ** 2) / 140],
            ]
        )

        Mb = Mb * rho * A * L

        Mc = np.array(
            [
                [1 / 3, 0, 0, 0, 0, 0],
                [0, 13 / 35, 0, 0, 0, -11 * L / 210],
                [0, 0, 13 / 35, 0, 11 * L / 210, 0],
                [0, 0, 0, I1 / (3 * A), 0, 0],
                [0, 0, 11 * L / 210, 0, L ** 2 / 105, 0],
                [0, -11 * L / 210, 0, 0, 0, L ** 2 / 105],
            ]
        )

        Mc = Mc * rho * A * L

        Ml = np.zeros(shape=(12, 12))
        Ml[0:6, 0:6] = Ma
        Ml[0:6, 6:] = Mb
        Ml[6:, 0:6] = Mb.T
        Ml[6:, 6:] = Mc

        fl = np.array(
            [
                [1 / 6 * L * (2 * f1[0] + f2[0])],
                [1 / 70 * L * (26 * f1[1] + 9 * f2[1])],
                [1 / 70 * L * (26 * f1[2] + 9 * f2[2])],
                [0],
                [-1 / 420 * (L ** 2) * (22 * f1[2] + 13 * f2[2])],
                [1 / 420 * (L ** 2) * (22 * f1[1] + 13 * f2[1])],
                [1 / 6 * L * (f1[0] + 2 * f2[0])],
                [1 / 70 * L * (9 * f1[1] + 26 * f2[1])],
                [1 / 70 * L * (9 * f1[2] + 26 * f2[2])],
                [0],
                [1 / 420 * (L ** 2) * (13 * f1[2] + 22 * f2[2])],
                [-1 / 420 * (L ** 2) * (13 * f1[1] + 22 * f2[1])],
            ]
        )

        return [Ml, K1, fl]
    
    @staticmethod
    def GetDofsLoop(nl_links, nl_flags, nl_links_properties, nl_alternates=None):
        DofsLoop = np.zeros(shape=(nl_links.shape[0] * 6, 11))
        count = 0

        for e in range(nl_links.shape[0]):
            link_nodes = nl_links[e, :]
            link_dofs = Assembly.elementglobaldofs(link_nodes)
            link_flags = nl_flags[e, :]

            if nl_alternates is not None:
                checkalt = np.isin(e, nl_alternates[:, 0])
                if checkalt:
                    check_pos = nl_alternates[:, 0].index(e)
            else:
                checkalt = False

            for d in range(6):
                DofsLoop[count, 0:2] = [link_dofs[d], link_dofs[d + 6]]
                DofsLoop[count, 2:] = nl_links_properties

                if link_flags[d] == 0:
                    # Change parameters of links to force linear behavior if degree
                    # of freedom is not activeted (a=1.0, AmpBW=0.0)
                    DofsLoop[count, 2], DofsLoop[count, 8] = 0.0
                elif checkalt:
                    # Change parameters of links if it is included in the different
                    # Bouc-Wen parameters matrix
                    DofsLoop[count, 2:] = nl_alternates[check_pos, 1:]

                count += 1

            # DofsLoop = DofsLoop.astype(int)
        return DofsLoop

    @staticmethod
    def elementglobaldofs(nodes):
        dofs = np.zeros(shape=(len(nodes) * 6))
        dofs[0 : len(nodes) * 6 : 6] = (nodes - 1) * 6
        dofs[1 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 1
        dofs[2 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 2
        dofs[3 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 3
        dofs[4 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 4
        dofs[5 : len(nodes) * 6 : 6] = (nodes - 1) * 6 + 5
        return dofs
```

#### Newmark simulation class

This class takes care of the time integration using the Newmark algorithm. It requires an Assembly object as an input, where the system matrices are defined, along with all necessary simulation parameters.

```python
class Newmark:
    def __init__(self, Assembly):
        self.Parameters = Assembly.SimulationParameters
        self.Assembly = Assembly
        self.InputFile = Assembly.InputFile

    def simulation(self, Assembly=None):
        if Assembly is None:
            Assembly = self.Assembly

        dt = self.Parameters.dt
        alphas = Newmark.SolverParams(dt=dt)
        a1, a2, a3, a4, a5, a6 = (
            alphas[0],
            alphas[1],
            alphas[2],
            alphas[3],
            alphas[4],
            alphas[5],
        )

        # Obtain system matrices
        Kstiff = Assembly.FinalMatrices["Kstiff"]
        M = Assembly.FinalMatrices["Mmass"]
        fint = Assembly.FinalMatrices["fint"]
        C = Assembly.FinalMatrices["Crayleigh"]
        
        #Correction to avoid singularities due to virtual links
        test = np.diag(M)
        test = test[test > 0]
        m = np.min(test) / 1000
        for i in range(np.shape(M)[0]):
            if M[i, i] == 0:
                M[i, i] = m

        a1M, a4C = a1 * M, a4 * C

        # Initialize
        DisplacementsU = Assembly.Displacements
        VelocitiesV = Assembly.Velocities
        AccelerationsA = Assembly.Accelerations
        HistDict = Assembly.HistDict

        HystereticCurvesR = HistDict["HystereticCurvesR"]
        HystereticCurvesU = HistDict["HystereticCurvesU"]

        v0 = VelocitiesV[:, 0]
        u0 = DisplacementsU[:, 0]

        # Extract external forcing
        fnorm = Assembly.fnorm
        Fmatrix = Assembly.Rmatrix

        # Compute initial conditions
        f0 = Fmatrix[:, 0]
        a0 = np.linalg.inv(M).dot((f0 - Kstiff.dot(u0) - C.dot(v0)))

        uk = u0
        uik = uk
        vk = v0
        ak = a0

        tol, maxit = 1e-03, 20
        # Start iterations
        for i in range(np.shape(Fmatrix)[1] - 1):
            fi = Fmatrix[:, i + 1]

            va1 = a2 * vk + a3 * ak
            va2 = a5 * vk + a6 * ak

            # Compute residual
            R = M.dot(-va1) + C.dot(va2) + fint[:, 0] - fi

            Rnorm = np.linalg.norm(R)
            nit = 0
            ui = uk
            uik = ui - uk

            while (Rnorm > tol * fnorm) and (nit <= maxit):
                Keff = a1M + a4C + Kstiff
                # Correct displacement based on residual
                du = np.linalg.inv(Keff).dot(-R)
                ui = ui + du
                DispsU = ui
                uik = ui - uk

                # Update system matrices due to nonlinearity
                globalMatrices = Assembly.assemble(DispsU, HistDict)
                FinalMatrices = Assembly.applyBCs(
                    Assembly.InputFile,
                    globalMatrices["Kstiff"],
                    globalMatrices["Mmass"],
                    globalMatrices["fint"],
                )

                HistDict = globalMatrices["Hist"]
                Kstiff = FinalMatrices["Kstiff"]
                M = FinalMatrices["Mmass"]
                fint = FinalMatrices["fint"]

                # Re-compute residual
                R = M.dot(a1 * uik - va1) + C.dot(a4 * uik + va2) + fint[:, 0] - fi
                Rnorm = np.linalg.norm(R)

                nit = nit + 1

            if nit <= maxit:
                if i % 500 == 50:
                    print(
                        "Step %i converged after %i iterations, residual %f \n"
                        % (i, nit, Rnorm / fnorm)
                    )
            else:
                print(
                    "Step %i did not converge after %i iterations, residual %f \n"
                    % (i, nit, Rnorm / fnorm)
                )

            DisplacementsU[:, i + 1] = DispsU
            uk = ui
            ak = a1 * uik - va1
            vk = a4 * uik + va2
            VelocitiesV[:, i + 1] = vk
            AccelerationsA[:, i + 1] = ak
            HystereticCurvesU[:, i + 1] = HistDict["HistUm"]  # HistDict["HistZeta"]
            HystereticCurvesR[:, i + 1] = HistDict["Rs"]

        Results = dict()
        Results["Disps"] = DisplacementsU
        Results["Velocities"] = VelocitiesV
        Results["AccelerationsA"] = AccelerationsA
        Results["HystereticParamsR"] = HystereticCurvesR
        Results["HystereticParamsU"] = HystereticCurvesU

        return Results

    def SolverParams(dt=1 / 100, gamma=1 / 2, beta=1 / 4):
        # Newmark parameters
        a1 = 1 / (beta * (dt ** 2))
        a2 = 1 / (beta * dt)
        a3 = (1 - 2 * beta) / (2 * beta)
        a4 = gamma / (beta * dt)
        a5 = 1 - gamma / beta
        a6 = (1 - gamma / (2 * beta)) * dt

        return [a1, a2, a3, a4, a5, a6]
```

#### Excitation definition

This class contains secondary functions needed to define the excitation signal. Several alternatives are provided, documented in detail in the corresponding *ReadMe* file. In short, the user is able to define the input ground motion signal as a multisine excitation, a white noise like excitation or pick from two example signals provided. For the first two alternatives the user is able to define the spectral components of the signal and the temporal amplitude coefficient. In addition, the user is able to define the angle of ground motion, however, currently only an in-plane angle definition is encoded in the notebook. 

```python
class SimulationParameters:
    def __init__(
        self,
        fs=100,             #Sampling frequency
        ICs=None,           #Initial Conditions
        angle=np.pi / 4,    #In-plane(x-y) angle of ground motion
        AmpF=10,            #Amplitude coefficient
        type="ExampleB",    #Type of excitation signal ("ExampleA","ExampleB","multisine","whitenoise")
        fminimum=3,         #Minimum frequency component of the signal in the multisine or white noise like cases
        fmax=10,            #Maximum frequency component of the signal in the multisine or white noise like cases
        Ncomponents=1000,   #Number of frequency components (multisine) or Number of samples (white noise)
        fstep=None,         #Variable to take every fstep frequencies to construct the multisine signal, if None the values is defined by the Ncomponents, if 
        seedrnd=30,         #Seed for the random white noise process
    ):

        # integration sampling frequency.
        self.dt = 1 / fs
        # integration time step.
        self.angle = angle
        self.ICs = ICs

        self.lowpass = False

        if type == "ExampleA":
            Excite = scipy.io.loadmat("ExampleA.mat")
            Rth = np.array(Excite.get("ExciteA"))
        elif type == "ExampleB":
            Excite = scipy.io.loadmat("ExampleB.mat")
            Rth = np.array(Excite.get("ExciteB"))
        elif type == "multisine":
            # Multisine constructor
            Rth = SimulationParameters.Multisine(fminimum, fmax, fs, Ncomponents, fstep)
        elif type == "whitenoise":
            # White Noise constructor
            Rth = SimulationParameters.WhiteNoise(
                fminimum, fmax, fs, samples=Ncomponents, seed=seedrnd
            )
            Rth = Rth / np.amax(abs(Rth))

        self.SynthesizedAccelerogram = AmpF * Rth
        
    @staticmethod
    def Multisine(fmin, fmax, fsample, Ncomponents=None, fstep=None):

        if fstep == 0:
            if Ncomponents is None:
                Ncomponents = 10
            vals = np.random.uniform(fmin, fmax + 1, Ncomponents)
        elif fstep is not None:
            vals = np.arange(fmin, fmax + 1, fstep)
        elif Ncomponents is not None:
            fstep = (fmax + 1 - fmin) / Ncomponents
            vals = np.arange(fmin, fmax + 1, fstep)
        else:
            vals = np.random.uniform(fmin, fmax + 1, 10)

        t = np.arange(0, 20 + 1 / fsample, 1 / fsample)
        x = np.zeros((len(t),))

        for k in vals:
            phiinit = random.uniform(0, 2 * np.pi)
            x += np.sin(2 * np.pi * k * t + phiinit)

        x = x / np.amax(abs(x)) * random.uniform(0.2, 1)

        return x

    @staticmethod
    def fftnoise(f, seedex=30):
        f = np.array(f, dtype="complex")
        Np = (len(f) - 1) // 2
        random.seed(seedex)
        phases = np.random.rand(Np) * 2 * np.pi
        phases = np.cos(phases) + 1j * np.sin(phases)
        f[1 : Np + 1] *= phases
        f[-1 : -1 - Np : -1] = np.conj(f[1 : Np + 1])

        return np.fft.ifft(f).real

    @staticmethod
    def WhiteNoise(min_freq, max_freq, fsample, samples=1024, seed=30):

        freqs = np.abs(np.fft.fftfreq(samples, 1 / fsample))
        f = np.zeros(samples)
        idx = np.where(np.logical_and(freqs >= min_freq, freqs <= max_freq))[0]
        f[idx] = 1

        return SimulationParameters.fftnoise(f, seed)
```

#### Main File

Main file for the evaluation of the response of the two story frame with Bouc-Wen links.

```python
# Define model Input

# Model definition - See Input File section
# Format: Mesh(lengthx, lengthy, heightz, type='All_Links')
Model = InputFile(7.50, 5.00, 3.2, "All_Links")
ndim = np.shape(Model.Nodes)[0] * 6

# Beam material definition - See Material Definition section
# Format:Material(Mesh, Eyoung=210e9, nee=0.30, rho=8000)
MaterialProperties = Material(Eyoung=210e9, nee=0.30, rho=8000)

# Define Bouc-Wen model parameters - See Bouc-Wen model section
# Format: BoucWenModel(Beta,Gamma,bwa=0.1,AmpBW=1.0,Alpha=1.0,N=1,deltav=0, deltan=0,bw_k=1.62e8)
Beta, Gamma = 6.0, 4.0
BoucWen = BoucWenModel(
    Beta, Gamma, AmpBW=1.0, bwa=0.1, bw_k=1.62e5, deltav=2.0, deltan=1.5,
)

# Define Initial Conditions
# (displacements/velocities/accelerations)
ICs = dict()
ICs["Disps"] = np.zeros((ndim, 1))
ICs["Velocities"] = np.zeros((ndim, 1))
ICs["Accelerations"] = np.zeros((ndim, 1))

# Damping parameters
# Rayleigh Damping with damping ratios (zetas) for the modes in the OmegaIndexes array
# See: Utils.GetRayleighDamping to change damping definition
ICs["zeta"] = [0.02, 0.02]
ICs["OmegaIndexes"] = [0, 1]

# Time integration parameters
Parameters = SimulationParameters(
    fs=100,
    ICs=ICs,
    AmpF=0.10,
    angle=np.pi / 4,
    type="ExampleB",
    fminimum=3,
    fmax=20,
    Ncomponents=2048,
    fstep=1,
    seedrnd=30,
)

# Assemble model
ModelAssembly = Assembly(Model, MaterialProperties, Parameters, BoucWen)

# Simulate analysis
NewmarkSolver = Newmark(ModelAssembly)
Results = NewmarkSolver.simulation()
```

```python
#Post-processing

# Results format:
# Results['Disps'] = Displacement response
# Results['Velocities'] = Velocity response
# Results['AccelerationsA'] = Acceleration response
# All these are matrices. Every row corresponds to a degree of freedom (spatial discretization)
# every column to a timestep (time integration)
# Results['HystereticParamsR'] = Hysteretic forcing matrix
# Results['HystereticParamsU'] = Hysteretic displacement matrix for plotting


# Plot Response
U = Results["Disps"]
# Pick dof
ndof_plot = np.argmax(np.amax(abs(U), axis=1))

# Plot
plt.plot(U[ndof_plot, :])
plt.xlabel("Time increments")
plt.ylabel("Displacement Response")
plt.tight_layout()
plt.show()

# Plot Hysteresis curve
Rh = Results["HystereticParamsR"]
Uh = Results["HystereticParamsU"]
# Pick dof
ndof_plot = np.argmax(np.amax(abs(Rh), axis=1))

# Plot
plt.plot(Uh[ndof_plot, :], Rh[ndof_plot, :])

plt.xlabel("Incremental Displacements")
plt.ylabel("Hysteretic Forcing")
plt.tight_layout()
plt.show()
```
